```python
import matplotlib.pyplot as plt
import numpy as np

def plot_loss(accuracy, train_loss,val_loss):
    #  accuracy
    plt.plot(accuracy)
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')
    plt.show()

    #  loss
    plt.plot(train_loss)
    plt.plot(val_loss)
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')
    plt.ylim(0, 30)
    plt.show()
```

```python

def logs_to_lists(execution_logs):
    train_losses = []
    valid_losses = []
    valid_accuracies = []
    
    execution_logs_list=list(execution_logs.rstrip().lstrip().splitlines())

    for epoch_line in execution_logs_list:
        a,b,c = epoch_line.split(' , ')
        train_losses.append(a.split(': ')[1])
        valid_losses.append(b.split(': ')[1])
        valid_accuracies.append(c.split(': ')[1])
        
    train_losses = [float(item) for item in train_losses]
    valid_losses = [float(item) for item in valid_losses]
    valid_accuracies = [float(item) for item in valid_accuracies]

    return train_losses, valid_losses, valid_accuracies
```
