## Geneate labels from permutations of two objects, two colors and two modifiers 

```python
import itertools
import src.alphabet as alphabet

SIMPLE_OBJECTS = ['lion', 'cross']
SIMPLE_COLORS = ['G', 'S'] # G: red , S: black
SIMPLE_MODIFIERS = ['rampant', 'dancetty']

all_possible_permutations = []

# one color & one object 
s1 = [alphabet.SIMPLE_COLORS, alphabet.SIMPLE_OBJECTS ]
permutations1 = list(dict.fromkeys(itertools.product(*s1)))

# one color & one object & one moddifier 
s2 = [alphabet.SIMPLE_COLORS, alphabet.SIMPLE_OBJECTS, alphabet.SIMPLE_MODIFIERS ]
permutations2 = list(dict.fromkeys(itertools.product(*s2)))


# two color & two objects 
s3 = [alphabet.SIMPLE_COLORS, alphabet.SIMPLE_COLORS, alphabet.SIMPLE_OBJECTS, alphabet.SIMPLE_OBJECTS ]
permutations3 = list(dict.fromkeys(itertools.product(*s3)))

# two color & two objects & two moddifier 
s4 = [alphabet.SIMPLE_COLORS, alphabet.SIMPLE_COLORS, alphabet.SIMPLE_OBJECTS, alphabet.SIMPLE_OBJECTS,  alphabet.SIMPLE_MODIFIERS, alphabet.SIMPLE_MODIFIERS ]
permutations4 = list(dict.fromkeys(itertools.product(*s4)))

all_possible_permutations = permutations1 + permutations2 + permutations3 + permutations4

print('Total number of permutations:', len(all_possible_permutations))
all_possible_permutations
```

```python
import random
from PIL import Image, ImageDraw
from pathlib import Path

COLORS_MAP = {
    'G': 'red',
    'S': 'black'
}

MODIFIERS_MAP = {
    'rampant': [(200,200), (300, 300), (100,20)],
    'dancetty': (200, 100, 300, 200)
}

def create_empty_image(im_path):
    image = Image.new("RGB", (400, 400), "white")
    image.save(im_path)
    return image

def image_exists(im_path):
    return Path(im_path).is_file()


def draw(shape, color, coordiantes, im_path): 
    if not image_exists(im_path):
        create_empty_image(im_path)
        
    with Image.open(im_path) as im:
        if shape == 'lion':
            im = draw_lion(im, coordiantes, color)
        elif shape == 'cross':
            im = draw_cross(im, coordiantes, color)

        im.save(im_path)
        
def draw_lion(im, coordiantes, color):
    draw = 
    .Draw(im)
    draw.polygon([(200,10), (200, 200), (150,50)], fill=color)
    
    return im

def draw_cross(im, coordiantes, color):
    draw = ImageDraw.Draw(im)
    draw.rectangle((200, 100, 300, 200), fill=color)

    return im


def get_coordinates(modifier):
    coordiantes = MODIFIERS_MAP.get(modifier)
    if coordiantes is None or coordiantes == '':
        coordiantes = (0,0)
 
    return coordiantes


def get_color(color_code):
    return COLORS_MAP.get(color_code)
    

```

```python
from src.label_checker_automata import LabelCheckerAutomata
automata = LabelCheckerAutomata()

FOLDER_NAME = 'generated-data'
counter = 1
for mutation in all_possible_permutations:
    label = ' '.join(mutation)
    parsed_label = automata.parse_label(label)
    output = automata.align_parsed_label(label, parsed_label)
    image_full_path = FOLDER_NAME + '/' + 'image_' + str(counter) + '.png'
    
    for obj, col in zip(output['objects'], output['colors']):
        draw(shape=obj, 
             color=get_color(col),
             coordiantes=(),#get_coordinates(col),
             im_path=image_full_path)
        print('Image "{}" for label "{}" has been generated succfully' .format(image_full_path, label) )
    counter = counter + 1

```
